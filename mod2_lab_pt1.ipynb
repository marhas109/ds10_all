#### Overview of Convolution Neural Networks

The data science team has chosen to use a Convolution Neural Network (CNN) approach to develop an automated image recognition system for quality inspections. CNNs are based on a deep learning algorithm specifically created to make image processing more effective and more efficient. They are able to autonomously extract features from images at a large scale, bypassing the need for manual input into feature development.

CNNs are built from a set of specialized layers:

1. Convolution layers apply mathematical functions called kernels to a matrix of pixels from the image. Each kernel moves across the entire image and recognizes a specific feature of the image: lines, curves, edges, shapes, etc.
2. Activation layers introduce non-linearity into the system, allowing the network to learn more complex patterns and ensuring the data flowing through the system remains in a calculable range.
3. Pooling layers reduce the dimensionality of the output of the other layers, while still preserving essential information about the image.
4. Fully connected layers integrate information about all features and produce the final output of the model.

#### How CNNs Process Image Data

CNN architecture is inspired by the similarly layered architecture of the human visual cortex. An important CNN concept ported directly from the biological inspiration is the _receptive field_.

- In the human visual cortex, the receptive field is the subset of the entire field of view that an individual neuron can access. The human visual cortex combines the data from the receptive field of each neuron to produce the complete image our brains "see."
- In a CNN, the receptive field is the size of the image region that produces a given feature; a CNN model combines data from each kernel's receptive field to create its interpretation of the entire image.

CNNs iterate over image data in each convolution layer. The initial convolution layers identify simple features (lines, edges). The processing of the early convolution layers allow each subsequent layer to have larger receptive fields, and therefore to identify increasingly complex features. In general, the middle convolution layers recognize image parts and the deepest layers identify complete objects. 

#### Description of a CNN Processing a Simple Image

For a simplified, conceptual explanation of how CNNs work, let's start with this simple image of a house:

In the early convolution layers of a CNN, the kernels would scan the image and identify things like the straight lines in the door, the curved lines in the window plants, and the texture of the siding.

In the middle convolution layers, the kernels would recognize complete shapes, such as the rectangles in the door and the triangle of the roof line/upper floor.

The deepest convolution layers would identify objects such as a door, windows, and a chimney. The CNN would ultimately be able to combine these elements and recognize the entire image as a house.
